```python
#Removal of Outliers
def remove_outlier(dataset,k=3.33):
    for col in dataset.columns:
            if (dataset[col].dtype=="int64" or dataset[col].dtype=="float64"):
                mean = dataset[col].mean()
                global ds      
                std = dataset[col].std()    
                outlier = [i for i in dataset[col] if (i > mean - k * std)]
                outlier = [i for i in outlier if (i < mean + k * std)]       
                ds = dataset.loc[dataset[col].isin(outlier)]
                

remove_outlier(data,k=3.33)
```

### TOP 10 PROFITABLE SELLING ITEMS

```python
items = data.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by = 'Profit',ascending = False)
items.head(10)
```

```python
plt.figure(figsize =(20,10))
sns.catplot('Sub-Category','Profit',data=items.head(10),kind='bar',aspect=2.5,height=8,palette='coolwarm')
plt.title('TOP 10 PROFITABLE PRODUCTS',size=25,color = 'Green')
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('CUMULATIVE PROFIT',size=15)
plt.xlabel('PRODUCTS',size=15)
```

### TOP 10 PROFITABLE CITIES

```python
city = data.groupby('City')['Profit'].sum().reset_index().sort_values(by = 'Profit',ascending = False)
city.head(10)
```

```python
plt.figure(figsize =(20,10))
sns.catplot('City','Profit',data=city.head(10),kind='bar',aspect=3,height=8,palette='cividis')
plt.title('TOP 10 CITIES WITH PROFITABLE',size=25,color = 'Green')
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('CUMULATIVE PROFIT',size=15)
plt.xlabel('CITY',size=15)
```
