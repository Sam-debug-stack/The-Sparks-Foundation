# Decision Tree Algorithm

### Predict the right class using Decision Tree classifier and visualize it graphically.

```python
# Importing libraries 
import pandas as pd
```

```python
#importing the Iris dataset with pandas
df = pd.read_csv(r"C:\Users\Gavas\Desktop\Iris.csv")
df
```

```python
df.head(5)
```

```python
df.describe()
```

```python
y=df['Species'].values
X=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
# Defining the decision tree algorithm

from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
#check the accuracy

pred=dt.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
print('Accuracy of the model is',accuracy_score(pred,y_test))
print(classification_report(pred,y_test))
```

###### visualize the Decision Tree

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
feature_cols=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

```python
# Visualize the graph

plt.figure(figsize=(50,30))

tree.plot_tree(dt.fit(X,y),feature_names=feature_cols,filled=True,precision=3,proportion=True,rounded=True)
plt.show()
```
